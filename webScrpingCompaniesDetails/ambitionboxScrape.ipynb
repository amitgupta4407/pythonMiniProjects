```python
headres ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
# add header file to avoid 403
URL = "https://www.ambitionbox.com/list-of-companies?page=1"
response = requests.get(URL, headers = headres).text
soup = BeautifulSoup(response, 'lxml')

# 30 companies per page, loop print anme of all of then
# for i in range(30):
#   print(soup.find_all("h2","company-name")[i].text.strip())

companies= soup.find_all("div",class_ = 'company-content-wrapper')
 
name = []
rating = []
review_count = []
company_type = []
company_old= []
company_loc = []
company_esiz = []
for i in companies:
  name.append(i.find('h2', class_ = "company-name" ).text.strip())
  rating.append(i.find("p", class_ = "rating").text.strip())
  review_count.append(i.find("a", class_ = "review-count").text.strip())
  try:
    company_type.append(i.find_all("p", class_ = "infoEntity")[0].text.strip())
  except:
    company_type.append(np.nan)
  try:
    company_loc.append(i.find_all("p", class_ = "infoEntity")[1].text.strip())
  except:
    company_loc.append(np.nan)
  try:
    company_old.append(i.find_all("p", class_ = "infoEntity")[2].text.strip())
  except:
    company_old.append(np.nan)
  try:
    company_esiz.append(i.find_all("p", class_ = "infoEntity")[2].text.strip())
  except:
    company_esiz.append(np.nan)
```

```python
name = []
rating = []
review_count = []
company_type = []
company_old= []
company_loc = []
company_esiz = []

def scrape_custom_page(page_no):
  headres ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

  URL = "https://www.ambitionbox.com/list-of-companies?page="+str(page_no)
  response = requests.get(URL, headers = headres)
  print(response)
  soup = BeautifulSoup(response.text, 'lxml')

  # 30 companies per page, loop print anme of all of then
  # for i in range(30):
  #   print(soup.find_all("h2","company-name")[i].text.strip())

  companies= soup.find_all("div",class_ = 'company-content-wrapper')
  
  for i in companies:
    name.append(i.find('h2', class_ = "company-name" ).text.strip())
    rating.append(i.find("p", class_ = "rating").text.strip())
    review_count.append(i.find("a", class_ = "review-count").text.strip())
    try:
      company_type.append(i.find_all("p", class_ = "infoEntity")[0].text.strip())
    except:
      company_type.append(np.nan)
    try:
      company_loc.append(i.find_all("p", class_ = "infoEntity")[1].text.strip())
    except:
      company_loc.append(np.nan)
    try:
      company_old.append(i.find_all("p", class_ = "infoEntity")[2].text.strip())
    except:
      company_old.append(np.nan)
    try:
      company_esiz.append(i.find_all("p", class_ = "infoEntity")[2].text.strip())
    except:
      company_esiz.append(np.nan)
for i in range(1,334):
  print(i+1)
  scrape_custom_page(i)
d={"name":name,
   "rating":rating,
   "review_count":review_count,
   "type":company_type,
   "how_old":company_old,
   "hq":company_loc,
   "employees":company_esiz
   }
df = pd.DataFrame(d)
df
```
